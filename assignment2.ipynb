There is an inverse direct correlation between reward and steps taken for this environment. This is because the value and rule for the reward is -1 per step.
<br>
Agents that have a higher learning rate(LR) seem to converge and perform better than agents with lower learning rates.
<br>
Agents that have a higher explortion factor(EF) have larger vallies. I believe this to be because a higher EF leads to more exploration and more penilties.
<br>
On average the agent with a higher exploration value will perform less effevtively.
<br>
There seeps to be 2 jumps that each agent takes and the severity seems to be correlated to the exploration rate. The first jump is at the start, there is a steady downwards trend then starts to move upward, the second is another upwards jump close to the end of training.
<br>
An agent that has a EF that is too large and too small can lead to some instability.
<br>
<br>
After reviewing the results I would propose the best agent would be (LR: 0.2, EF: 0.1). The high learning rate will converge fast and a middle ground value for exploration factor will be stable.

```python
from assignment2_utils import *
from agent_handler import Agent_handler
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(results, result_index, window_size, title):
  plt.figure(figsize=(25, 10))

  for result in results:
      sns.lineplot(np.convolve(result[result_index], np.ones(window_size) / window_size, mode='same'), label=f"(LR, EF) {(result[2]) }")
      
  plt.title(title)
  plt.legend()
  plt.show()
```

```python
env = gym.make('Taxi-v3')

num_obs, num_actions = describe_env(env)

agent_handler = Agent_handler({
  "num_episodes": 2000,
  "num_obs": num_obs,
  "num_actions": num_actions,
  "learning_rates": [0.01, 0.001, 0.2],
  "discount_factor": 0.9,
  "exploration_factors": [0.2, 0.3]
})

results1 = agent_handler.train(env)
results1 = [(t[1], t[2], t[3]) for t in results1]
```

#### __Note: In the following plots I have implemented some smoothing to the graphs so we can actually gleam some meaningful insight from them.__

```python
smoothing = 15
plot(results1, 0, smoothing, "Number of steps per episode")
```
