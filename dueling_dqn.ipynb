```python
import gymnasium as gym
from dqn_agent import DQN_Agent
from agent_handler import Agent_handler as AgentHandler
from dueling_dqn.agent import DuelingDQN_Agent
from dueling_dqn.utils import describe_env, plot_results
```

#### 1. Initialize Pacman Environment

```python
env = gym.make("MsPacman-v4")
num_obs, num_actions = describe_env(env)
```

```python
handler = AgentHandler(
    {"num_episodes": 20, "max_steps": 10000, "notify_percent": 5, "update_rate": 10}
)
```

#### 2. Train DQN Agent for comparison

```python
dqn_agent = DQN_Agent(
    {
        "num_obs": num_obs,
        "num_actions": num_actions,
        "learning_rate": 0.01,
        "discount_factor": 0.9,
        "exploration_factor": 0.2,
    }
)

results = handler.train([dqn_agent], env)
plot_results(results, 2, 1, "DQN Agent")
```

#### 3. Train Dueling DQN Agent

```python
dueling_dqn_agent = DuelingDQN_Agent(
    {
        "num_obs": num_obs,
        "num_actions": num_actions,
        "learning_rate": 0.05,
        "discount_factor": 0.9,
        "exploration_factor": 0.2,
    }
)
results = handler.train([dueling_dqn_agent], env)
plot_results(results, 2, 1, "Dueling DQN Agent")
```
